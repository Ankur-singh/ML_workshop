# Student clustering

Every child is not the same. They all have different requirements, different ways of learning and speed. So I don't single study pattern for all the different student is correct. Its a very important job to identify different sets of student, understand their differences and respect them.

For a human it might be a difficult task if the number of student are more. Machine Learning is at your rescue! 

## Importing Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib notebook
```

##  Load and Plotting Data  

```python
data = pd.read_csv('Datasets/student-por.csv',sep = ';')
data.head()
```

```python
print(data.info())
```

```python
data.shape
```

##  Data Preprocessing

```python
data1=pd.get_dummies(data,drop_first = True)
data1.head()
```

##  Dimesionalty Reduction

```python
X2 = PCA(3).fit_transform(data1)
X2
```

```python
fig = plt.figure(figsize=(10,8))
ax = Axes3D(fig)
ax.scatter(X2[:, 0], X2[:, 1],X2[:,2], s=[100]*len(X2))
```

## Unsupervised Clustering

```python
y_pred = KMeans(n_clusters=3).fit(data1)
labels = y_pred.predict(data1)
```

```python
fig = plt.figure(figsize=(10,8))
ax = Axes3D(fig)
ax.scatter(X2[:, 0], X2[:, 1],X2[:,2], s=[100]*len(X2),c=labels, cmap='rainbow')
```
