# Fermont bridge data analysis

The Fremont Bridge Bicycle Counter records the number of bikes that cross the bridge using the pedestrian/bicycle pathways. Inductive loops on the east and west pathways count the passing of bicycles regardless of travel direction. The data consists of a date/time field: Date, east pathway count field: Fremont Bridge NB, and west pathway count field: Fremont Bridge SB. The count fields represent the total bicycles detected during the specified one hour period

## Importing Libraries

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt 
plt.style.use('seaborn')
import seaborn as sns
%matplotlib inline
```

## Getting Data

```python
data = pd.read_csv('Datasets/Fremont.csv',index_col='Date',parse_dates=True)
data.head()
```

```python
data.columns=['East','West']
data.resample('W').sum().plot(figsize=(15,8));
```

```python
data['Total'] = data['West']+data['East']
ax=data.resample('D').sum().rolling(365).sum().plot(figsize=(15,8))
ax.set_ylim(0,None);
```

```python
data.groupby(data.index.time).mean().plot(figsize=(15,8));
```

```python
pivoted = data.pivot_table('Total',index = data.index.time,columns = data.index.date)
pivoted.iloc[:5,:5]
```

```python
pivoted.plot(legend = False,alpha = 0.01,figsize=(15,8));
```

## <span style="color:Blue">Prinicipal Component Analysis

```python
X = pivoted.fillna(0).T.values
X.shape
```

```python
from sklearn.decomposition import PCA
X2 = PCA(2, svd_solver='full').fit_transform(X)
X2.shape
```

```python
plt.figure(figsize=(15,8))
plt.scatter(X2[:, 0], X2[:, 1]);
```

## <span style="color:Blue">Unsupervised Clustering

```python
gmm = GaussianMixture(2).fit(X)
labels = gmm.predict(X)
```

```python
plt.figure(figsize=(17,8))
plt.scatter(X2[:, 0], X2[:, 1], c=labels, cmap='rainbow')
plt.colorbar();
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

pivoted.T[labels == 0].T.plot(legend=False, alpha=0.1, ax=ax[0]);
pivoted.T[labels == 1].T.plot(legend=False, alpha=0.1, ax=ax[1]);

ax[0].set_title('Purple Cluster')
ax[1].set_title('Red Cluster');
```

## <span style="color:Blue">Comparing With Days Of Week

```python
dayofweek = pd.DatetimeIndex(pivoted.columns).dayofweek
plt.figure(figsize=(17,8))
plt.scatter(X2[:, 0], X2[:, 1], c=dayofweek, cmap='rainbow')
plt.colorbar();
```
